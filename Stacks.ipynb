### Stack with size limit

```python
class Stack:
    def __str__(self, maxSize):
        self.maxSize = maxSize
        self.list = []
        
    def __str__(self):
        values = [str(x) for x in reversed(self.list)]
        return '\n'.join(values)
    
    def isEmpty(self):
        if self.list == []:
            return True
        else:
            return False
        
    # isFull method, will return True or False depending if stack is full
    # Only works with Stack with limited size
    def isFull(self):
        if len(self.list) == maxSize:
            return 'List is full'
        else:
            return False
        
    def push(self, value):
        if self.isFull:
            return 'Stack is full'
        else:
            self.list.append()
            return 'The appended value has been added'
        
        def pop(self):
            if isEmpty():
                return 'List is empty'
            else:
                return self.list.pop()
                    
        def peek(self):
            if isEmpty():
                return 'List is empty'
            else:
                return self.list.[len(self.list-1])
            
        def delete(self):
            self.list = None
        
                                
    
```
